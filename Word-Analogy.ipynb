```python
!pip install gensim
from gensim.models import KeyedVectors
```

```python
word_vectors = KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True)
```

```python
def find_analogies(word1, word2, word3):
    # word1 - word2 = ? - word3
    # e.g king - main = ? - woman
    # ? = +king +woman -man
    r = word_vectors.most_similar(positive=[word1, word3], negative=[word2])
    print("%s - %s = %s - %s" % (word1, word2, r[0][0], word3))
```

```python
find_analogies("king", "man", "woman")
```

```python
find_analogies("france", "paris", "london")
```

```python
find_analogies("france", "paris", "rome")
```

```python
# nearest neighbors
def nearest_neighbors(word):
    r = word_vectors.most_similar(positive=[word])
    print("neighbors of: %s" % word)
    for word, score in r:
        print("\t%s" % word)
```

```python
nearest_neighbors("father")
```

```python
nearest_neighbors("wife")
```
