Labelling train data as 0 and 1 for each poet

```python
input_texts = []
labels = []
for label, f in enumerate(import_files):
  print(f"{f} corresponds to label {label}")
  for line in open(f):
    line = line.rstrip().lower()
    if line:
      line = line.translate(str.maketrans('', '', string.punctuation))
      input_texts.append(line)
      labels.append(label)
```

Taking out train data and test data

```python
train_text, test_text, Ytrain, Ytest = train_test_split(input_texts, labels)
```

Setting index as 1 and storing unkown words in 0 index

```python
idx = 1
word2idx = {"<unk>": 0}
```

```python
for text in train_text:
  token = text.split()
  for word in token:
    if word not in word2idx:
      word2idx[word] = idx
      idx += 1
```

converting the words in word2idx into numbers

```python
train_text_int = []
test_text_int = []

for text in train_text:
  tokens = text.split()
  line_as_int = [word2idx[token] for token in tokens]
  train_text_int.append(line_as_int)

for text in test_text:
  tokens = text.split()
  line_as_int = [word2idx.get(token, 0) for token in tokens]
  test_text_int.append(line_as_int)
```

using Bayes classification model to compute A and pi for markov model

```python
V = len(word2idx)

A0 = np.ones((V, V))
pi0 = np.ones(V)

A1 = np.ones((V, V))
pi1 = np.ones(V)


def compute_counts(text_as_int, A, pi):
  for tokens in text_as_int:
    last_idx = None
    for idx in tokens:
      if last_idx is None:
        pi[idx] += 1
      else:
        A[last_idx, idx] += 1

      last_idx = idx

  return A, pi
```

```python
compute_counts([t for t, y in zip(train_text_int, Ytrain) if y == 0], A0, pi0)
```

```python
compute_counts([t for t, y in zip(train_text_int, Ytrain) if y == 1], A1, pi1)
```

```python
A0 /= A0.sum(axis=1, keepdims=True)
pi0 /= pi0.sum()

A1 /= A1.sum(axis=1, keepdims=True)
pi1 /= pi1.sum()
```

```python
logA0 = np.log(A0)
logpi0 = np.log(pi0)

logA1 = np.log(A1)
logpi1 = np.log(pi1)
```

```python
count0 = sum(y == 0 for y in Ytrain)
count1 = sum(y == 1 for y in Ytrain)
total = len(Ytrain)
p0 = count0 / total
p1 = count1 / total
logp0 = np.log(p0)
logp1 = np.log(p1)
p0, p1
```

Creating the classifier model

```python
class Classifier:
  def __init__(self, logAs, logpis, logpriors):
    self.logAs = logAs
    self.logpis = logpis
    self.logpriors = logpriors
    self.K = len(logpriors)

  def _compute_log_likelihood(self, input_, class_):
    logA = self.logAs[class_]
    logpi = self.logpis[class_]

    last_idx = None
    logprob = 0
    for idx in input_:
      if last_idx is None:
        logprob += logpi[idx]
      else:
        logprob += logA[last_idx, idx]
      last_idx = idx

    return logprob

  def predict(self, inputs):
    predictions = np.zeros(len(inputs))
    for i, input_ in enumerate(inputs):
      posteriors = [self._compute_log_likelihood(input_, c) + self.logpriors[c] \
                     for c in range(self.K)]
      pred = np.argmax(posteriors)
      predictions[i] = pred
    return predictions
```

```python
clf = Classifier([logA0, logA1], [logpi0, logpi1], [logp0, logp1])
```

```python
Ptrain = clf.predict(train_text_int)
print(f"Train acc: {np.mean(Ptrain == Ytrain)}")
```

```python
Ptest = clf.predict(test_text_int)
print(f"Test acc: {np.mean(Ptest == Ytest)}")
```
