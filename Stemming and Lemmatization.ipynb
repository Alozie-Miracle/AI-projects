```python
from nltk.corpus import wordnet
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("walking")
```

```python
lemmatizer.lemmatize("walking", pos=wordnet.VERB)
```

```python
lemmatizer.lemmatize("mice")
```

```python
def get_wordnet_pos(treebank_tag):
  if treebank_tag.startswith("J"):
    return wordnet.ADJ
  elif treebank_tag.startswith("V"):
    return wordnet.VERB
  elif treebank_tag.startswith("N"):
    return wordnet.NOUN
  elif treebank_tag.startswith("R"):
    return wordnet.ADV
  else:
    return wordnet.NOUN
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
sentence = "Donald Trump has a devoted following".split()
sentence
```

```python
from nltk.tag import pos_tag
```

```python
words_and_tags = pos_tag(sentence)
words_and_tags
```

```python
for words, tags in words_and_tags:
  print(lemmatizer.lemmatize(words, pos=get_wordnet_pos(tags)))
```

```python
sentence = "The cat was following the bird as it flew by".split()
```

```python
words_and_tags = pos_tag(sentence)
words_and_tags
```

```python
for word, tags in words_and_tags:
  print(lemmatizer.lemmatize(word, pos=get_wordnet_pos(tags)), end=" ")
```
