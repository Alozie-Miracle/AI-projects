```python
current_index = 0
```

```python
word2idx = {}
```

```python
documents = ['I like cats', "I love cats", "I love dogs"]
```

```python
for doc in documents:
  tokens = doc.split()
  for word in tokens:
    if word not in word2idx:
        word2idx[word] = current_index
        current_index += 1
```

```python
print(word2idx)
```
